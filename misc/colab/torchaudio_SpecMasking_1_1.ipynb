```python
import torch, torchaudio, IPython
from matplotlib import pyplot as plt

audio_path = '/Users/eldadoh/PycharmProjects/environmental-classifier/_data_local/Assets/sensi-env-clf-benchmark/dataset_samples_by_classes/toilet/task_id_62558_annotation_2_AS_20220730-100926-0-29960-3reality-00E93AAC9BCB.ogg'
waveform, fs = torchaudio.load(audio_path)
```

```python
fbank = torchaudio.compliance.kaldi.fbank(waveform, htk_compat=True, sample_frequency=16000, use_energy=False,
                                          window_type='hanning', num_mel_bins=128, dither=0.0, frame_shift=10)

freqm = torchaudio.transforms.FrequencyMasking(48)
timem = torchaudio.transforms.TimeMasking(1000)

fbank = torch.transpose(fbank, 0, 1)
```

```python
waveform.shape, fbank.shape
IPython.display.Audio(waveform, rate=16000)
```

```python
# NOTE this line, this is the trick, new torchaudio expect [1, freq, time] while old support [freq, time]
# comment this line will lead to an issue.
fbank = fbank.unsqueeze(0)
print(fbank.shape)
fbank = freqm(fbank)

plt.imshow(fbank[0].cpu().numpy())

fbank = timem(fbank)

plt.imshow(fbank[0].cpu().numpy())
fbank = fbank.squeeze(0)
print(fbank.shape)
```
